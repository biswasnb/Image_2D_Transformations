```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

#upload and load image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

```

```python
#show function
def show(img,title):
  plt.imshow(img)
  plt.title(title)
  plt.axis("off")
  plt.show()

show(img,"Original Image")

```

```python
#Transformations
rows,cols=img.shape[:2]

#Translation
trans_matrix=np.float32([[1,0,100],[0,1,100]])
translated =cv2.warpAffine(img,trans_matrix,(cols,rows))
show(translated,"Translation")
```

```python
#Rotation
rot_matrix=cv2.getRotationMatrix2D((cols//2,rows//2),45,1)
rotated =cv2.warpAffine(img,rot_matrix,(cols,rows))
show(rotated,"Rotation")
```

```python

#Scaling
scaled=cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_LINEAR)
show(scaled,"Scaling")
```

```python
#Shearing
shear_matrix=np.float32([[1, 0.2, 0], [0.2, 1, 0]])
sheared=cv2.warpAffine(img,shear_matrix,(cols,rows))
show(sheared,"Shearing")
```

```python
#Reflection
reflected_matrix=np.float32([[-1, 0, cols], [0, 1, 0]])
reflected=cv2.warpAffine(img,reflected_matrix,(cols,rows))
show(reflected,"Reflection")
```
